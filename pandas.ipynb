```python
#Using Dictionaries
Subject_marks={
    'maths':100,
    'science':99,
    'english':98,
    'social':97,
    'cn':86,
    'dbms':75,
    'toc':50
}
Subject_marks
```

```python
s2.loc['science']
```

```python
#conditional selection it will give boolean values
s2>99
```
